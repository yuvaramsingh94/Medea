#### PubMed

```python
pubmed = PubMedAPI(
    top_k_results=5,   # number of papers to fetch
    email="your_email@example.com"  # recommended by PubMed
)
```

```python
articles = pubmed.load("diabetes treatment")

for article in articles:
    print(article["Title"])
    print(article["Published"])
    print(article["Abstract"])
    print(article["url"])
    print("-" * 50)
```

```python
for article in pubmed.lazy_load("cancer immunotherapy"):
    print(article["Title"])
    print(article["Abstract"])
    print(article["url"])
    print(article["Published"])
    print("-" * 50)
```

```python
print("\n=== DICT TO DOCUMENT ===")
doc = pubmed._dict2document(articles[0])
print(doc.page_content)   # abstract
print(doc.metadata)       # rest of fields
```

```python
docs = pubmed.load_docs("diabetes treatment")

for d in docs:
    print(d.page_content[:100])  # abstract preview
    print(d.metadata)
```

```python
query = "recent treatments for diabetes"

item = {
    "input": query,
    "ctxs": []   # we will fill this
}
```

```python
papers = search_paper_via_query(query, max_paper_num=5)

ctxs = []
if papers:
    for p in papers:
        if p.get("abstract"):
            ctxs.append({
                "text": p["abstract"],
                "title": p.get("title", ""),
                "citation_counts": p.get("citationCount", 0)
            })

item["ctxs"] = ctxs
```

```python
papers = search_paper_via_query("Diabetes", max_paper_num=3, attempt=5)

for p in papers:
    print("TITLE:", p["title"])
    print("ABSTRACT:", p["abstract"])
    
    # Try getting full text if ArXiv
    if p.get("externalIds") and "ArXiv" in p["externalIds"]:
        passages = ArxivPaperProcessor.retrieve_passages(p["externalIds"]["ArXiv"])
        print("FULL TEXT SAMPLE:", passages[:2])  # first 2 paragraphs
    
    print("-" * 50)
```
